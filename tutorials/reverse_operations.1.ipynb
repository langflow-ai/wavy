```python
import pandas as pd
import yfinance as yf

import wavy
from wavy.utils import reverse_pct_change, reverse_diff
```

```python
tickers = pd.read_html('https://en.wikipedia.org/wiki/List_of_S%26P_500_companies')[0].Symbol
```

```python
msft = yf.Ticker("MSFT")
```

```python
# get historical market data from 2005 on
df = msft.history(period="max", start="2005-01-01")
```

```python
df = df[['Close']]
```

```python
df.head()
```

```python
x, y = wavy.create_panels(df, lookback=10, horizon=1)
```

```python
### Fist Difference
```

```python
x_diff = x.diff_()[1:]
y_diff = y.diff_()[1:]
```

```python
model = wavy.LinearRegression(x_diff, y_diff)
```

```python
y_pred = model.predict()
```

```python
y_true = y.as_dataframe()
y_reversed = reverse_diff(y_pred.as_dataframe(), y_true)
```

```python
pd.DataFrame({'y_true': y_true.Close,'y_pred': y_reversed.Close}).plot()
```

```python
### Percent Change
```

```python
x_change = x.pct_change_()[1:]
y_change = y.pct_change_()[1:]
```

```python
model = wavy.LinearRegression(x_change, y_change)
```

```python
y_pred = model.predict()
```

```python
y_true = y.as_dataframe()
y_reversed = reverse_pct_change(y_pred.as_dataframe(), y_true)
```

```python
pd.DataFrame({'y_true': y_true.Close,'y_pred': y_reversed.Close}).plot()
```
