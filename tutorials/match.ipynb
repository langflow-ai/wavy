```python
import pandas as pd
import yfinance as yf

import wavy
```

```python
tickers = pd.read_html('https://en.wikipedia.org/wiki/List_of_S%26P_500_companies')[0].Symbol
```

```python
msft = yf.Ticker("MSFT")
```

```python
# get historical market data from 2005 on
df = msft.history(period="max", start="2005-01-01")
```

```python
df = df[['Close']]
df.head()
```

```python
# Create panels to predict one step into the future
x, y = wavy.create_panels(df, lookback=10, horizon=1)
```

```python
y.plot()
```

```python
# pct_change at panel-level
y_change = y.pct_change_()
```

```python
# y_change[0] is nan
y_change[0]
```

```python
# Remove NaN values created by pct_change
y_change = y_change[1:]
```

```python
# End up w/ different lengths for x and y_change
len(y_change), len(x)
```

```python
# Matching to the rescue!
x = x.match(y_change)
```
